Preprocessing Data

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/PancreaticCancerData.csv')
df = df.set_index('sample_id')

Xdf = df.loc[:, ['age', 'sex','plasma_CA19_9','creatinine','LYVE1','REG1B','TFF1','REG1A']]
Xdf = Xdf.fillna(value=0)
Ydf = df.diagnosis

L=len(Ydf)
for i in range(L):
    if Ydf[i]==3:
        Ydf[i]=1
    else:
        Ydf[i]=0

Xdf=pd.get_dummies(Xdf)
```

Train - Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(Xdf,
                                                    Ydf,
                                                    test_size=0.2,
                                                    train_size=0.8,
                                                    stratify=Ydf,
                                                    random_state=42)
```

Support Vector Machine

```python
from sklearn import svm
from sklearn import metrics

clf = svm.SVC(kernel='linear') 
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```
